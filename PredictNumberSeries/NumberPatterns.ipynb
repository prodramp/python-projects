# **Objective:**
- Generate a sequence of numbers based on Function Y
- Y = A Function which is based on x
- x is a series of numbers based on uniform number distribution


```python
n = 100
x = np.random.uniform(size=(n,1))  
y = np.sqrt(x)
```

Adding noice to the function
- Random disturbance to the series

```python
n = 1000
x = np.random.uniform(size=(n,1))  
noise = np.random.normal(0, 0.01, size=x.shape)
y = np.sqrt(x)
y = y + noise
```

```python
fig, ax = plt.subplots(dpi=100, figsize=(15,5))
plt.scatter(x,y.ravel())
```

Using Functions

```python
# Equation:: y = x*x
def generate_data_with_noise(n, is_noise):
  x = np.random.uniform(size=(n,1))  
  y = np.square(x) 
  if is_noise:
    noise = np.random.normal(0, 0.1, size=x.shape)
    y = y + noise
  return (x, y)
```

```python
def generate_numbers_plot(x,y):
  fig, ax = plt.subplots(dpi=100, figsize=(15,5))
  plt.scatter(x,y.ravel())
```

```python
x,y = generate_data_with_noise(100, True)
generate_numbers_plot(x,y)
```

# **Plot to show both graphs side by side**

```python
def side_by_side_plots(title, x1,y1, x2,y2):
  fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15,5))
  fig.suptitle(title)
  ax1.scatter(x1, y1.ravel())
  ax2.scatter(x2, y2.ravel())
```

# Equation:: y = -x*x - x - 1

```python
# Equation:: y = -x*x - x - 1
def generate_data_with_noise_2(n, is_noise):
  x = np.random.uniform(size=(n,1))  
  y = -np.square(x) - x - 1
  if is_noise:
    # reduce range from 0 to 0.01 to make the sharper curve
    noise = np.random.normal(0, 0.1, size=x.shape)
    y = y + noise
  return (x, y)
```

```python
x1,y1 = generate_data_with_noise_2(100, False)
x2,y2 = generate_data_with_noise_2(100, True)
side_by_side_plots('-x*x - x - 1', x1,y1, x2,y2)
```

```python
# 
# Equation:: y = x²
def generate_data_with_noise_3(n, is_noise):
  x = np.random.uniform(size=(n,1))  
  y = np.square(x) - (x + 1)
  if is_noise:
    # reduce range from 0 to 0.01 to make the sharper curve
    noise = np.random.normal(0, 0.01, size=x.shape)
    y = y + noise
  return (x, y)
```

```python
x1,y1 = generate_data_with_noise_3(100, False)
x2,y2 = generate_data_with_noise_3(100, True)
side_by_side_plots('x*x - (x + const)', x1,y1, x2,y2)
```

```python
# 
# Equation:: -x² + x + 2
def generate_data_with_noise_4(n, is_noise):
  x = np.random.uniform(size=(n,1))  
  y = -np.square(x) + x + 1
  if is_noise:
    # reduce range from 0/1 to 0.1/0.01 to make the sharper curve
    noise = np.random.normal(0, 0.01, size=x.shape)
    y = y + noise
  return (x, y)
```

```python
x1,y1 = generate_data_with_noise_4(100, False)
x2,y2 = generate_data_with_noise_4(100, True)
side_by_side_plots('-x*x - x - 1', x1,y1, x2,y2)
```

# **Exponential Function**

```python
# -----
# Equation:: np.exp(x)
def generate_exponential_data_with_noise(n, is_noise):
  # Add range from 0 to 1/10
  x = np.random.uniform(0, 10, size=(n,1))  
  y = np.exp(x)
  if is_noise:
    # reduce range from 0/1 to 100/1000 to make the sharper curve
    noise = np.random.normal(0, 1000, size=x.shape)
    y = y + noise
  return (x, y)
```

```python
x1,y1 = generate_exponential_data_with_noise(100, False)
x2,y2 = generate_exponential_data_with_noise(100, True)
side_by_side_plots('Exponential Curve', x1,y1, x2,y2)
```

# **Logarithmic Function**

```python
# -----
# Equation:: np.log(x)
def generate_logarithmic_data_with_noise(n, is_noise):
  # Add range from 0 to 1/10
  x = np.random.uniform(0, 10, size=(n,1))  
  y = np.log(x)
  if is_noise:
    # reduce range from 0/1 to 100/1000 to make the sharper curve
    noise = np.random.normal(0, 0.1, size=x.shape)
    y = y + noise
  return (x, y)
```

```python
x1,y1 = generate_logarithmic_data_with_noise(100, False)
x2,y2 = generate_logarithmic_data_with_noise(100, True)
side_by_side_plots('Logarithmic Curve (np.log(x))', x1,y1, x2,y2)
```

# **Trignometric Functions (Sin, Cos, Tan)**

```python
# Trignometric - Sin/Cos/Tan > Sin
# Sin > np.sin(2 * np.pi * x)
def generate_sin_with_noise(n, is_noise):
  x = np.random.uniform(size=(n,1))  
  y = np.sin(2*np.pi*x)
  if is_noise:
    # reduce range from 0/1 to 0.1/0.01 to make the sharper curve
    noise = np.random.normal(0, 0.1, size=x.shape)
    y = y + noise
  return (x, y)
```

```python
x1,y1 = generate_sin_with_noise(100, False)
x2,y2 = generate_sin_with_noise(100, True)
side_by_side_plots('Sin Curve', x1,y1, x2,y2)
```

```python
# Trignometric - Sin/Cos/Tan > Cos
# Cos > np.cos(2 * np.pi * x)
def generate_cos_with_noise(n, is_noise):
  x = np.random.uniform(size=(n,1))  
  y = np.cos(2*np.pi*x)
  if is_noise:
    # reduce range from 0/1 to 0.1/0.01 to make the sharper curve
    noise = np.random.normal(0, 0.1, size=x.shape)
    y = y + noise
  return (x, y)
```

```python
x1,y1 = generate_cos_with_noise(100, False)
x2,y2 = generate_cos_with_noise(100, True)
side_by_side_plots('Cos Curve', x1,y1, x2,y2)
```

```python
# Trignometric - Sin/Cos/Tan > Tan
# Tan > np.arctan(2 * np.pi * x)
def generate_tan_with_noise(n, is_noise):
  x = np.random.uniform(size=(n,1))  
  y = np.arctan(2*np.pi*x)
  if is_noise:
    # reduce range from 0/1 to 0.1/0.01 to make the sharper curve
    noise = np.random.normal(0, 0.1, size=x.shape)
    y = y + noise
  return (x, y)
```

```python
x1,y1 = generate_tan_with_noise(100, False)
x2,y2 = generate_tan_with_noise(100, True)
side_by_side_plots('Tan Curve', x1,y1, x2,y2)
```
