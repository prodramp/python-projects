```python
from tensorflow.keras.applications.resnet50 import ResNet50
from tensorflow.keras.layers import Input
```

```python
input_shape = (512,512,3)
```

```python
inputs = Input(input_shape)
```

```python
base_model_nc = ResNet50(weights='imagenet', include_top=True, input_tensor=inputs)
```

```python
base_model_nc.summary()
```

```python
base_model_nc.save('resnet100.h5')
```
