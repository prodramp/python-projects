```python
from tensorflow.keras.applications.resnet50 import ResNet50
from tensorflow.keras.layers import Input
```

### Adding fully connected layers: ###
- include_top=True => include the fully-connected layer at the top of the network
- Neurons in a fully connected layer have full connections to all activations in the previous layer, as seen in regular Neural Networks

```python
input_shape = (512,512,3)
```

```python
inputs = Input(input_shape)
```

```python
base_model_nc = ResNet50(weights='imagenet', include_top=False, input_tensor=inputs)
```

```python
base_model = ResNet50(weights='imagenet', input_tensor=inputs)
```

```python
import numpy as np
from keras.preprocessing.image import image
import matplotlib.pyplot as plt
from tensorflow.keras.applications.resnet50 import preprocess_input
from keras.applications.imagenet_utils import decode_predictions
```

```python
img = image.load_img('bicycle-kid.jpg', target_size = input_shape)
plt.imshow(img)
img = image.img_to_array(img)
img = np.expand_dims(img, axis=0)
```

```python
img = preprocess_input(img)
```

```python
preds = base_model.predict(img)
```

```python
preds_nc = base_model_nc.predict(img)
```

```python
preds_nc
```

```python
preds
```

```python
decode_predictions(preds_nc, top=5)
```
