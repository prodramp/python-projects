```python
import pandas as pd
```

```python
aisles = pd.read_csv('aisles.csv')
```

```python
departments = pd.read_csv('departments.csv')
```

```python
products = pd.read_csv('products.csv')
```

```python
aisles
```

```python
departments
```

```python
products
```

```python
prod_with_aisles = pd.merge(products, aisles, how='inner', on="aisle_id")
```

```python
prod_with_aisles
```

```python
prod_final = pd.merge(prod_with_aisles, departments, how='inner', on="department_id")
```

```python
prod_final
```

```python
prod_final.columns
```

```python
prod_final = prod_final[['product_id', 'product_name', 'aisle', 'department']]
```

```python
prod_final.to_csv('products_list.csv')
```
