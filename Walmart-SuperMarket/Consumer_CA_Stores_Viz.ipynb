```python
import pandas as pd
```

```python
ca_purchase_history = "https://github.com/prodramp/python-projects/raw/main/Walmart-SuperMarket/purchase_history_ca.csv"
```

```python
stores_url = "https://github.com/prodramp/python-projects/raw/main/Walmart-SuperMarket/master_store_data.csv"
```

```python
stores_with_state = 'https://github.com/prodramp/python-projects/raw/main/Walmart-SuperMarket/stores_with_state_geo.csv'
```

```python
user_address = 'https://github.com/prodramp/python-projects/raw/main/Walmart-SuperMarket/user_address.csv'
```

```python
pur_his = pd.read_csv(ca_purchase_history)
```

```python
pur_his
```

```python
all_stores = pd.read_csv(stores_url)
```

```python
all_stores
```

```python
df_store_st = pd.read_csv(stores_with_state)
```

```python
df_store_st
```

```python
df_store_st = df_store_st[df_store_st['state'] == 'CA']
```

```python
df_store_st
```

```python
store_final = pd.merge(df_store_st, all_stores, how ='left' , on='name')
```

```python
store_final
```

```python
store_final = store_final[['name', 'store_id', 'latitude_x', 'longitude_x']]
```

```python
store_final
```

```python
store_final.columns = ['name', 'store_id', 'latitude', 'longitude']
```

```python
store_final
```

```python
pur_his
```

```python
pur_his = pur_his[['family_id', 'super_market_uuid']]
```

```python
pur_his
```

```python
df_shopping = pur_his.groupby(['family_id', 'super_market_uuid']).size().to_frame(name='total').reset_index()
```

```python
df_shopping
```

```python
consumer_address_df = pd.read_csv(user_address)
```

```python
consumer_address_df
```

```python
user_final_df = pd.merge(df_shopping, consumer_address_df, how='left', on='family_id')
user_final_df
```

```python
user_final_df = user_final_df[['family_id', 'super_market_uuid', 'total', 'longitude', 'latitude']]
```
