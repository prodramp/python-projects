```python
import pandas as pd
```

```python
master_store = 'https://github.com/prodramp/python-projects/raw/main/Walmart-SuperMarket/master_store_data.csv'
```

```python
ca_purchase_history = 'https://github.com/prodramp/python-projects/raw/main/Walmart-SuperMarket/purchase_history_ca.csv'
```

```python
user_address = 'https://github.com/prodramp/python-projects/raw/main/Walmart-SuperMarket/user_address.csv'
```

```python
stores_with_geo = 'https://github.com/prodramp/python-projects/raw/main/Walmart-SuperMarket/stores_with_state_geo.csv'
```

```python
purchase_ca = pd.read_csv(ca_purchase_history)
```

```python
purchase_ca
```

```python
all_stores = pd.read_csv(master_store)
```

```python
all_stores
```

```python
purchase_ca = purchase_ca[['family_id', 'super_market_uuid']]
```

```python
purchase_ca
```

```python
final_purchase_df = purchase_ca.groupby(['family_id', 'super_market_uuid']).size().to_frame(name='total').reset_index()
```

```python
final_purchase_df
```

```python
df_address = pd.read_csv(user_address)
```

```python
df_address
```

```python
add_temp_df = pd.merge(final_purchase_df, df_address, how='left', on='family_id')
add_temp_df
```

```python
purchase_df = add_temp_df[['family_id', 'super_market_uuid', 'total', 'longitude', 'latitude']]
purchase_df
```
